```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("historical_data.csv")
```

```python
# Heat map
df_without_strings = df.drop(["created_at","actual_delivery_time","store_primary_category","store_id","market_id","order_protocol","total_onshift_dashers","total_busy_dashers","total_outstanding_orders","estimated_order_place_duration","estimated_store_to_consumer_driving_duration"],axis=1)
CorrTable = df_without_strings.corr()
sns.heatmap(CorrTable, cmap = 'coolwarm')
```

```python
df = df[df['min_item_price']<12000]
plt.scatter(df["min_item_price"],df["subtotal"],alpha=0.5)
plt.xlabel("min_item_price")
plt.ylabel("subtotal")
plt.show()
```

```python
df = df[df['max_item_price']<12000]
plt.scatter(df["max_item_price"],df["subtotal"],alpha=0.5)
plt.xlabel("max_item_price")
plt.ylabel("subtotal")
plt.show()
```

```python
plt.scatter(df["num_distinct_items"],df["subtotal"],alpha=0.5)
plt.xlabel("distinct_items")
plt.ylabel("subtotal")
plt.show()
```

```python
df[df['total_items']>400]
df = df.drop(df[df['total_items']>400].index)
df[df['subtotal']>25000]
df = df.drop(df[df['subtotal']>25000].index)
plt.scatter(df["total_items"],df["subtotal"],alpha=0.2)
plt.xlabel("total_items")
plt.ylabel("subtotal")
plt.show()
```
